```python
# TASK 2 CALCULATOR

# Making a function for addition . . .
def addition(n1,n2):
    return n1+n2   

# Making a function for subtraction . . .
def subtraction(n1,n2):
    return n1-n2

# Making a function for multiplication . . .
def multiplication(n1,n2):
    return n1*n2

# Making a function for division . . .
def division(n1,n2):
    return n1/n2
    
# Asking the user to enter the first number . . .
while True:
    
    try:
        num1 = float(input("Enter the first number: "))
        print("Your first number is:", num1)
        break # we will simply break out of the loop if the number entered is valid . . .
        
    except ValueError:
        print("Invalid input! Please enter a number.")

# Asking the user to enter the first number . . .
while True:
    
    try:
        num2 = float(input("Enter the second number: "))
        print("Your second number is:", num2)
        break # we will simply break out of the loop if the number entered is valid . . .
    
    except ValueError:
        print("Invalid input! Please enter a number.")

# Asking the user which operation they want to perform on their two numbers . . .
operation = input("Enter the operation you wish to perform. (+,-,*,/)")
print("The user wishes to perform:",operation)

if operation == '+':
    # calling the function for addition and passing our two numbers as parameters . . .
    result=addition(num1,num2)
    print("The addition of two numbers is:",result)

elif operation == '-':
    # calling the function for subtraction and passing our two numbers as parameters . . .
    result=subtraction(num1,num2)
    print("The subtraction of two numbers is:",result)

elif operation == '*':
    # calling the function for multiplication and passing our two numbers as parameters . . .
    result=multiplication(num1,num2)
    print("The multiplication of two numbers is:",result)

elif operation == '/':
    
    # we gotta be careful, since we cannot divide by 0 . . .
    if num2 == 0:
        print("We cannot divide by 0")
    
    else:
        # calling the function for division and passing our two numbers as parameters . . .
        result=division(num1,num2)
        print("The division of two numbers is:",result)
        
else:
    print("Sorry! The operation you wish to perform is not available on our calculator.")
```

```python
# TASK 3 PASSWORD GENERATOR
import string
import random

# Making a function that will randomly generate a password based on characters (random) . . .
def generate_password(length,include_digits,include_uppercase,include_lowercase,include_punctuation):
    
    # Making a variable called characters . . .
    characters = ''
    
    # if it is digit, first converting it to lower and then checking if the promt was y or yes in that case we add it . . .
    if include_digits.lower() in ['y','yes']:
        characters += string.digits
        
    # if it is uppercase, we are checking if the promt was y or yes in that case we add it . . .    
    if include_uppercase.lower() in ['y','yes']:
        characters += string.ascii_uppercase

    # if it is lowercase, we are checking if the promt was y or yes in that case we add it . . .    
    if include_lowercase.lower() in ['y','yes']:
        characters += string.ascii_lowercase
      
    # if it is punctuation, we are checking if the promt was y or yes in that case we add it . . .    
    if include_punctuation.lower() in ['y','yes']:
        characters += string.punctuation
    
    # Check if at least one type of character is selected . . .
    if not characters:
        return "Error: No character types selected for password generation!"
    
    # now randomly generating the password . . .
    return ''.join(random.choice(characters)for i in range(length))

# input validity check . . .
while True:
    
    try:
        # prompting the user to enter the length of the password . . .
        length_password = int(input("Enter the length of your password"))
        print("The length of user's password is:",length_password)
        break
    except ValueError:
        print("Invalid input! Please enter a number.")
    
# To make it more interesting, I will ask the user if they want uppercase/lowercase or digits in their password etc . . .
include_digits = input("Do you wish to have digits in the password? (y/n)")
include_uppercase = input("Do you wish to have uppercase characters in the password? (y/n)")
include_lowercase = input("Do you wish to have lowercase characters in the password? (y/n)")
include_punctuation = input("Do you wish to have punctuations in the password? (y/n)")

# asking if they want digits in their password . . .
if include_digits == 'y' or include_digits == 'yes':
    print("User wants to include digits in their randomly generated password")

else:
    print("User do not want digits in their password.")
   
# asking if they want uppercase characters in their password . . .
if include_uppercase == 'y' or include_uppercase == 'yes':
    print("User wants to include uppercase characters in their randomly generated password")

else:
    print("User do not want uppercase characters in their password.")
    
# asking if they want lowercase characters in their password . . .
if include_lowercase == 'y' or include_lowercase == 'yes':
    print("User wants to include lowercase characters in their randomly generated password")

else:
    print("User do not want lowercase characters in their password.")
    
# asking if they want punctuations in their password . . .
if include_punctuation == 'y' or include_punctuation == 'yes':
    print("User wants to include punctuations in their randomly generated password")

else:
    print("User do not want punctuations in their password.")

# Generating the password by using a combiynation of random characters . . .
# calling a function and passing the length etc to it . . .

print("The randomly generated password is:",generate_password(length_password,include_digits,include_uppercase,include_lowercase,include_punctuation))
```

```python
# TASK 4 ROCK PAPER OR SCISSORS . . .
import random

# Making a function that will simulate the game being played . . .
def match():
    
    # First lets take the input from the user about his choice (rock,paper,scissor)
    choice_user = input("Rock, paper or scissors? ").lower()
    print("The user chose:",choice_user)
    
    # Now, randomly generating the choice for computer based on our list . . .
    choices = ["rock","paper","scissor"]
    
    # choice for computer . . .
    choice_computer = random.choice(choices)
    print("The computer chose:",choice_computer)
    
    # Now, let's determine who wins the game . . .
    if choice_user == choice_computer:
        print("Tie") # in this case it is a tie . . .
        
        # cases where user wins the match . . .
    elif (choice_user == "rock" and choice_computer == "scissor") or \
         (choice_user == "scissor" and choice_computer == "paper") or \
         (choice_user == "paper" and choice_computer == "rock"):
        return "user"
    
    else:
        return "computer" # this means that the computer won . . .

# Now, let us keep track of their respective scores . . .
user_score = 0
computer_score = 0

# since we want to give the option for multiple games being played . . .
while True:
    
    res = match() # match function called . . .
    
    # updating the scores . . .
    if res == "user": # indicates that the user won . . .
        user_score += 1
        print("The user has won this game.")
    
    # indicating that the computer won the game and score is increased . . .
    elif res == "computer":
        computer_score += 1
        print("Computer won the game.")
    
    # indicating that no one won . . .
    else:
        print("Tie!")

    # now lets print the score . . .
    print("The score of the user is:",user_score)
    print("The score of the computer is:",computer_score)
    
    # Now, lets give the option to the user if they want to play again . . .
    option = input("Do you want to play again? (y/n)").lower()
    if option !='y':
        break
        
# Final outcome, determining who is the winner . . .
if user_score > computer_score:
    print("The user has won the game")

elif user_score < computer_score:
    print("The computer has won the game")
    
else:
    print("The game is a tie.")    
```
